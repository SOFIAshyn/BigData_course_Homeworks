```python
import pandas as pd
import csv
import time
from datetime import datetime, timedelta
from collections import Counter
```

```python
# TASK 1
def get_list_of_all_accounts(lst_all_messages_data):
    return [account_dict['account'] for account_dict in lst_all_messages_data]
```

```python
def add_tweet_data(dict_of_collected_data, flag):
    if flag is True:
        dict_of_collected_data['tweets'] = []
    
    hashtags = []
    for word in message[6].split():
        if word.startswith('#'):
            hashtags.append(word.replace('#', ''))
    
    dict_of_collected_data['tweets'].append({
        'date': message[0],
        'text': message[6],
        'hashtags': hashtags
    })
```

```python
file = open('./training.1600000.processed.noemoticon.csv', 'r', encoding='latin-1')
csv_reader = csv.reader(file)
lst_all_messages_data = []

for message in csv_reader:
    message.insert(0, str(datetime.now()))
#     time.sleep(1)
    dict_of_collected_data = {}

#     message = (message.value).decode('utf-8').split(',')
    if message[5] not in set(get_list_of_all_accounts(lst_all_messages_data)):
        dict_of_collected_data['account'] = message[5]
        dict_of_collected_data['counts'] = 1
        add_tweet_data(dict_of_collected_data, flag=True)
        lst_all_messages_data.append(dict_of_collected_data)
    else:
        for dict_of_collected_data in lst_all_messages_data:
            if dict_of_collected_data['account'] == message[5]:
                dict_of_collected_data['counts'] += 1
                add_tweet_data(dict_of_collected_data, flag=False)
    
```

```python
time_frame_start = datetime.now() - timedelta(hours=3)
time_frame_end = datetime.now()

posts_lst_in_3_hours = []
dict_account_num_tweets_in_3_hours = {}

for account_dict in lst_all_messages_data:
    count_acount_tweets_in_3_hours = 0
    
    date_lst = [tweet['date'] for tweet in account_dict['tweets']]
    for date in date_lst:
        date = datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
        if time_frame_start < date and date < time_frame_end:
            count_acount_tweets_in_3_hours += 1
    dict_account_num_tweets_in_3_hours = {
        'account': account_dict['account'],
        'counts_3': count_acount_tweets_in_3_hours
    }
    
    posts_lst_in_3_hours.append(dict_account_num_tweets_in_3_hours)

```

```python
# TASK 2
dict_10_most_active_users_in_3_hours = []
sorted_posts_lst_in_3_hours = sorted(posts_lst_in_3_hours, key=lambda x: x['counts_3'], reverse=True)
users_most_posts_in_3_hours = [el['account'] for el in sorted_posts_lst_in_3_hours[:10]]
for user in users_most_posts_in_3_hours:
    for user_data in lst_all_messages_data:
        if user_data['account'] == user:
            user_10_last_posts = sorted(user_data['tweets'], key=lambda x: x['date'], reverse=True)[:10]
            dict_10_most_active_users_in_3_hours.append({
                'account': user,
                '10_last_tweets': user_10_last_posts
            })
```

```python
# RESULT DICT FOR TASK 2
{'10_most_active_users_in_3_hours': dict_10_most_active_users_in_3_hours}
```

```python
# TASK 3
# Show the aggregated statistics with a number of tweets per each hour for all the accounts (for the last 3 hours).

def choose_one_hour_tweets(sorted_user_posts, time_frame_start, time_frame_end):
#     date = datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
    count = 0
    posts_in_timeframe = []
    for tweets_section in sorted_user_posts:
        date = datetime.strptime(tweets_section['date'], '%Y-%m-%d %H:%M:%S.%f')

        if time_frame_start < date and date < time_frame_end:
            count += 1
        posts_in_timeframe.append(tweets_section)
    return {
        'tweets': posts_in_timeframe, 
        'count': count
    }
            
    

users_who_posted_in_3_hours = [el['account'] for el in sorted_posts_lst_in_3_hours]
lst_of_3_last_hours_statistic = []

for user in users_who_posted_in_3_hours:
    for user_data in lst_all_messages_data:
        if user_data['account'] == user:
            sorted_user_posts = sorted(user_data['tweets'], key=lambda x: x['date'], reverse=True)
            
            time_frame_start = datetime.now() - timedelta(hours=1)
            time_frame_end = datetime.now()
            one_hour_ago = choose_one_hour_tweets(sorted_user_posts, time_frame_start, time_frame_end)
            
            time_frame_start = datetime.now() - timedelta(hours=2)
            time_frame_end = datetime.now() - timedelta(hours=1)
            two_hour_ago = choose_one_hour_tweets(sorted_user_posts, time_frame_start, time_frame_end)
            
            time_frame_start = datetime.now() - timedelta(hours=3)
            time_frame_end = datetime.now() - timedelta(hours=2)
            three_hour_ago = choose_one_hour_tweets(sorted_user_posts, time_frame_start, time_frame_end)
            
            lst_of_3_last_hours_statistic.append({
                'account': user,
                'one_hour_ago_tweets': one_hour_ago,
                'two_hour_ago_tweets': two_hour_ago,
                'three_hours_ago_tweets': three_hour_ago
            })
```

```python
# RESULT DICT FOR TASK 3
{'3_last_hours_statistic': lst_of_3_last_hours_statistic}
```

```python
# TASK 4
# Show Top 20 most “tweet-producing” accounts for the last n hours (n should be given as a param).
N = 5
time_frame_start = datetime.now() - timedelta(hours=N)
time_frame_end = datetime.now()

in_n_hours_most_productive = []

for account_dict in lst_all_messages_data:
    count_acount_tweets_in_N_hours = 0

    date_lst = [tweet['date'] for tweet in account_dict['tweets']]
    for date in date_lst:
        date = datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
        if time_frame_start < date and date < time_frame_end:
            count_acount_tweets_in_N_hours += 1

    dict_acount_tweets_in_N_hours = {
        'account': account_dict['account'],
        'counts_N': count_acount_tweets_in_N_hours
    }

    in_n_hours_most_productive.append(dict_acount_tweets_in_N_hours)

```

```python
# RESULT DICT FOR TASK 4
{'in_n_hours_20_most_productive_accounts': \
 [dict_['account'] for dict_ in sorted(in_n_hours_most_productive, key=lambda x: x['counts_N'], reverse=True)[:20]]}
```

```python
# TASK 5
# Show the most popular hashtags across all the tweets during the last n hours (n should be given as a param)
N = 6
time_frame_start = datetime.now() - timedelta(hours=N)
time_frame_end = datetime.now()

# most_popular_hashtags = {}
all_hashtags = []

for account_dict in lst_all_messages_data:
    date_lst = [ ]
    for tweet in account_dict['tweets']:
        date = tweet['date']
        date = datetime.strptime(date, '%Y-%m-%d %H:%M:%S.%f')
        if time_frame_start < date and date < time_frame_end:
            all_hashtags += tweet['hashtags']
```

```python
# RESULT DICT FOR TASK 5
{'most_popular_hashtags_in_N_hours': \
 dict(sorted((Counter(all_hashtags)).items(), key=lambda item: item[1], reverse=True))}
```
